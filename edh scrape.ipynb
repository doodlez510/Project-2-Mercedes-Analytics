```python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import pandas as pd
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.mbofedh.com/used-vehicles/'
browser.visit(url)
```

```python
html=browser.html
```

```python
edh_soup = bs(html, 'html.parser')
```

```python
print(edh_soup.prettify())
```

```python
mylist=[]
options=edh_soup.find_all('div',class_='options')
for option in options:
    for li in option.find_all('li'):
            mydict={}
            mydict[li.text.split(':')[0]]=li.text.split(':')[1]
            details=option.text
            mylist.append(mydict)
```

```python
mylist2=[{k.replace('\n','').replace('\t',''): v.replace('\n','').replace('\t','') for k,v in x.items()} for x in mylist]
mylist2
pd.DataFrame(mylist2)
```

```python
cars=edh_soup.find_all('div',class_='vehicle-overview')
```

```python
stocks=[]
prices=[]
vins=[]
sale_prices=[]
titles=[]
years=[]

for car in cars:
    if car.find('div',class_='price-block our-price ') is not None:
        stock=car.find('span',class_='stock-label').text
        stock=stock.split(':')[1]
        stocks.append(stock)

        vin=car.span.text
        vin=vin.split(':')[1]
        vins.append(vin)
        if car.find('div',class_='price-block original') is not None:
            price=car.find('div',class_='price-block original').find('span',class_='price').text
            prices.append(price)
            prices=[x.replace('\n','').replace('\t','') for x in prices]

        sale_price=car.find('div',class_='price-block our-price ').find('span',class_='price').text
        sale_prices.append(sale_price)
        sale_prices=[x.replace('\n','').replace('\t','') for x in sale_prices]
        
        year=car.find('a').text
        if year.split(" ")[0] == "Certified":
            year=year.split(" ")[2]
            years.append(year)
        else:
            year=year.split(" ")[1]
            years.append(year)
        
        title=car.find('a').text
        if title.split(' ',2)[0]=='Pre-Owned':
            title=title.split(' ',2)[2]
            titles.append(title)
        else:
            title=title.split(' ',3)[3]
            titles.append(title)
```

```python
edh_df=pd.DataFrame({'Stock #': stocks,
'Original Price': prices,
'Sales Price': sale_prices,
'Vin #': vins,
'Vehicle Name':titles
})
edh_df
```
