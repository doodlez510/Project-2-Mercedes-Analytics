```python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import pandas as pd
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.mbofrocklin.com/used-vehicles/'
browser.visit(url)
```

```python
html=browser.html
```

```python
rocklin_soup = bs(html, 'html.parser')
```

```python
print(rocklin_soup.prettify())
```

```python
cars=rocklin_soup.find_all('div',class_='vehicle-overview')
```

```python
first_car=rocklin_soup.find('div',class_='vehicle-overview')
```

```python
vin_num=first_car.span.text
vin_num.split(':')[1]
```

```python
org_first_car_price=first_car.find('div',class_='price-block original')
org_first_car_price.find('span',class_='price').text
```

```python
sale_first_car_price=first_car.find('div',class_='price-block our-price ')
sale_first_car_price.find('span',class_='price').text
```

```python
vehicles=rocklin_soup.find_all('div',class_='vehicle-overview')
print(type(vehicles))
print(len(vehicles))
```

```python
mylist=[]
options=rocklin_soup.find_all('div',class_='options')
for car in options:
#     car_sublist = []
    mydict={}
    for li in car.find_all('li'):
        
        mydict[li.text.split(':')[0]]=li.text.split(':')[1]
        details=car.text
#         print(details, '-'*80)
    mylist.append(mydict)
mylist
```

```python
mylist2=[{k.replace('\n','').replace('\t',''): v.replace('\n','').replace('\t','') for k,v in x.items()} for x in mylist]
# mylist2 = [car_dict for car_dict in mylist]
mylist2
```

```python
options_df=pd.DataFrame(mylist2)
options_df
```

```python
stocks=[]
prices=[]
vins=[]
sale_prices=[]
titles=[]

for car in cars:
    if car.find('div',class_='price-block original') is not None:
        stock=car.find('span',class_='stock-label').text
        stock=stock.split(':')[1]
        stocks.append(stock)

        vin=car.span.text
        vin=vin.split(':')[1]
        vins.append(vin)

        price=car.find('div',class_='price-block original').find('span',class_='price').text
        prices.append(price)
        prices=[x.replace('\n','').replace('\t','') for x in prices]

        sale_price=car.find('div',class_='price-block our-price ').find('span',class_='price').text
        sale_prices.append(sale_price)
        sale_prices=[x.replace('\n','').replace('\t','') for x in sale_prices]
        
        title=car.find('a').text
        titles.append(title)
```

```python
rocklin_df=pd.DataFrame({'Stock #': stocks,
'Original Price': prices,
'Sales Price': sale_prices,
'Vin #': vins,
'Vehicle Name': titles
})
rocklin_df
```

```python
rocklin_df=pd.merge(rocklin_df,options_df,left_index=True,right_index=True)
```

```python
first_car=rocklin_soup_1.find('div',class_='vehicle-overview')
cars=rocklin_soup_1.find_all('div',class_='vehicle-overview')
```

```python
stocks=[]
prices=[]
vins=[]
sale_prices=[]
titles=[]
years=[]

for car in cars:
    if car.find('div',class_='price-block original') is not None:
        stock=car.find('span',class_='stock-label').text
        stock=stock.split(':')[1]
        stocks.append(stock)

        vin=car.span.text
        vin=vin.split(':')[1]
        vins.append(vin)

        price=car.find('div',class_='price-block original').find('span',class_='price').text
        prices.append(price)
        prices=[x.replace('\n','').replace('\t','') for x in prices]

        sale_price=car.find('div',class_='price-block our-price ').find('span',class_='price').text
        sale_prices.append(sale_price)
        sale_prices=[x.replace('\n','').replace('\t','') for x in sale_prices]
        
        year=car.find('a').text
        if year.split(" ")[0] == "Certified":
            year=year.split(" ")[2]
            years.append(year)
        else:
            year=year.split(" ")[1]
            years.append(year)
        
        title=car.find('a').text
        if title.split(' ',2)[0]=='Pre-Owned':
            title=title.split(' ',2)[2]
            titles.append(title)
        else:
            title=title.split(' ',3)[3]
            titles.append(title)
```

```python
rocklin_df_1=pd.DataFrame({'Stock #': stocks,
'Original Price': prices,
'Sales Price': sale_prices,
'Vin #': vins,
'Vehicle Name':titles,
'Year':years
})
rocklin_df_1
```

```python
rocklin_df_1=pd.merge(rocklin_df_1,options_df,left_index=True,right_index=True)
rocklin_df_1.head()
```
